# [Q.1] Datasets
Extract the datasets provided by the authors (ETT, ECL and Weather) and visualize each dataset in the form of a plot.

```

import pandas as pd
import matplotlib.pyplot as plt
ett_data = pd.read_csv('data/ETT/ETTh1.csv')


ett_data['date'] = pd.to_datetime(ett_data['date'])

plt.figure(figsize=(10, 6))


plt.plot(ett_data['date'], ett_data['OT'], label='OT')


plt.xlabel('Date')
plt.ylabel('Values')
plt.title('ETT Data')
plt.legend()
plt.grid(True)
plt.show()
```
